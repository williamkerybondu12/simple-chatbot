```python
def chatbot_response(user_input):
    user_input = user_input.lower()  # make input case-insensitive

    # Basic greetings
    if user_input == "hello":
        return "Hi there!"
    elif user_input == "hi":
        return "Hello!"
    elif user_input == "how are you":
        return "I'm doing great! How about you?"
    elif user_input == "what is your name":
        return "I'm a simple chatbot ."
    elif user_input == "what can you do":
        return "I can chat with you, answer simple questions, and keep you company!"
    elif user_input == "who created you":
        return "I was created by someone learning Python!"

    # Jokes
    elif user_input == "tell me a joke":
        return "Why did the computer get cold? Because it forgot to close its Windows!"
    elif user_input == "another joke":
        return "Why don't robots get tired? Because they recharge!"

    # Weather (general)
    elif user_input == "how's the weather":
        return "I can't check real weather, but I hope it's nice where you are!"
    elif user_input == "is it raining":
        return "I'm not sure, but keeping an umbrella nearby is always smart!"

    # Hobbies
    elif user_input == "what are your hobbies":
        return "I like chatting and helping people learn!"
    elif user_input == "suggest a hobby":
        return "How about reading books, learning guitar, painting, or coding?"

    # Fun conversation
    elif user_input == "do you like humans":
        return "Of course! You're fun to talk to."
    elif user_input == "what is your favorite color":
        return "I like White Color!"
    elif user_input == "tell me a fact":
        return "Did you know? Honey never spoils — archaeologists found 3000-year-old honey still good!"

    # Goodbye
    elif user_input == "bye":
        return "Goodbye! Have a great day!"
    else:
        return "I don't understand that. Try asking something else!"

def run_chatbot():
    print("Chatbot is running! Type 'bye' to exit.")

    while True:
        user_input = input("William: ")
        response = chatbot_response(user_input)
        print("Bot:", response)

        if user_input.lower() == "bye":
            break

# Start the chatbot
run_chatbot()
```
